```python
import cv2
import os
import face_recognition

# Define the input directory containing video files
input_directory = r'D:\dfd\dfdc project\Model Creation\celeborg'

# Define the output directory where you want to save the cropped videos
output_directory = r'D:\dfd\dfdc project\Model Creation\Celeb_Face_only_data1'

# Create the output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# List all video files in the input directory
video_files = [f for f in os.listdir(input_directory) if f.endswith('.mp4')]

# Loop through each video file and process it
for video_file in video_files:
    input_video_path = os.path.join(input_directory, video_file)

    # Open the input video file
    cap = cv2.VideoCapture(input_video_path)

    # Get video properties
    frame_width = int(cap.get(3))
    frame_height = int(cap.get(4))
    frame_rate = int(cap.get(5))

    # Define the codec and create a VideoWriter object for the output video
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    output_video_path = os.path.join(output_directory, f'{video_file}')
    out = cv2.VideoWriter(output_video_path, fourcc, frame_rate, (frame_width, frame_height))

    # Loop through frames and process them
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Find face locations in the frame
        face_locations = face_recognition.face_locations(frame)

        # Crop and process each face in the frame
        for face_location in face_locations:
            top, right, bottom, left = face_location
            face_frame = frame[top:bottom, left:right]

            # Write the face frame to the output video
            out.write(face_frame)

    # Release the input and output video objects for the current video
    cap.release()
    out.release()

# Print a message when processing is complete
print("All videos processed and saved with cropped faces.")
```
